### This is the code for Figure 4 from PeakTK paper
*Note that there are some changes from the example in the paper

```python
# import peaktk
import peaktk.dataloader.ISONE_helper
import peaktk.preprocessing.preprocessing
import peaktk.preprocessing.helper
import peaktk.demand_prediction.lstm_dp
import peaktk.peakday_prediction_monthly.vpeak
import peaktk.stats.peakday_helper
import peaktk.metrics.metrics
```

```python
# load ISOâˆ’NE data
DEMAND_PATH = "../data/ISONE/"
WEATHER_PATH = "../data/ISONE/weather"
ISONE_data = peaktk.dataloader.ISONE_helper.load_data(DEMAND_PATH,WEATHER_PATH)
```

```python
# data preprocessing and preparation
ISONE_data_scale, y_scaler = peaktk.preprocessing.preprocessing.scale(ISONE_data, min=0 , max=1) 
X, y, idx = peaktk.preprocessing.preprocessing.create_dataset(ISONE_data_scale) 
X_train, X_test, y_train, y_test, idx_train, idx_test= peaktk.preprocessing.helper.train_test_split(X, y, idx, test_size =366)
```

### Predict demand using LSTMDemandPrediction()

```python
# train demand prediction
lstm_model = peaktk.demand_prediction.lstm_dp.LSTMDemandPrediction() 
lstm_model.fit(X_train, y_train, epochs=1000, verbose =1)
# predict peak demand
y_hat = lstm_model.predict(X_test)
pred_demand = y_scaler.inverse_transform(y_hat).flatten()
```

### Predict Peak Day of the Month

#### VPeak Example

```python
# set the number of peak days to predict and delta day
num_peakday, delta = 3, 3

# prepare monthly data for training VPeak
historical_load, prev_year_load = peaktk.preprocessing.helper.prepare_vpeak_training_data(ISONE_data, current_year=2020)

# create and train peak day prediction model from historical load
# IMPORTANT NOTE: YOU SHOULD PERFORM HYPERPARAMETER TUNING AT THIS STEP.
vpeak_model = peaktk.peakday_prediction_monthly.vpeak.VPeak_Selection() 
vpeak_model.train(historical_load, prev_year_load, top_k=num_peakday+delta, quantile=0.9)
# predict peak day one day at a time
y_pred_peak = []
for i in range (len(pred_demand)):
    curr_date = idx_test[i]
    curr_pred_demand = pred_demand[i]
    is_peak = vpeak_model.predict_ispeak(curr_date, curr_pred_demand)
    y_pred_peak.append(is_peak)
```

```python
# find groundtruth peak day
# y_test_peak = peaktk.stats.peakday_helper.identify_monthly_peak(ISONE_data, top_k=num_peakday)
y_test_peak = peaktk.stats.peakday_helper.find_monthly_peak(y_test, idx_test, top_k=num_peakday)

# evaluate the results
precision = peaktk.metrics.metrics.precision(y_test_peak, y_pred_peak)
recall = peaktk.metrics.metrics.recall(y_test_peak, y_pred_peak)
balanced_accuracy = peaktk.metrics.metrics.balanced_accuracy(y_test_peak, y_pred_peak)
```
